\vec{h} = \begin{bmatrix} 3 \\3 \\ 3 \\ 3 \end{bmatrix}

\begin{align}
A = \begin{bmatrix}

B = \begin{bmatrix}

\end{bmatrix}

C = \begin{bmatrix}

D = \begin{bmatrix}

\end{bmatrix}
\qquad
E = \begin{bmatrix}
1 & 3 \\
5 & 7
\end{bmatrix}
\end{align}

```
A = np.matrix([[1, 2], [3, 4], [5, 6]])
B = np.matrix([[2, 4, 6]])
C = np.matrix([[9, 6, 3], [4, 7, 11]])
D = np.matrix([[1, 0, 0], [0, 1, 0], [0, 0, 1]])
E = np.matrix([[1, 3], [5, 7]])

mat_list = [A, B, C, D, E]

for i in mat_list:
    print(f'The shape of {i} is {i.shape}')
    
'''
All with the same inner numbers, so AC, AE can work, BA, BD can work, CA, CD can work, DA can work, and EC can work
'''
```

## 2.2 Find the following products: CD, AE, and BA. What are the dimensions of the resulting matrices? How does that relate to the dimensions of their factor matrices?

```
CD = np.matmul(C, D)
AE = np.matmul(A, E)
BA = np.matmul(B, A)

print(f'CD total is {CD}, dimensions are {CD.shape}.\nAE total is {AE}, dimensions are {AE.shape}.\nBA total is {BA}, dimensions are {BA.shape}.\n')

'''
When multipling matrices, the number of rows the result will be comes from the number of rows from the first matrix, and the number of columns comes from the number of columns in the second matrix.

'''
```

## 3.2 Find $|H|$ and then find $|J|$.

\begin{align}
H= 
\begin{bmatrix}

\end{bmatrix}
\qquad
J= 
\begin{bmatrix}

```
H = np.matrix([[12, 11],
             [7, 10]])

J = np.matrix([[0, 1, 2],
             [7, 10 ,4],
             [3, 2, 0]])

print(np.linalg.det(H))
print(np.linalg.det(J))
```
